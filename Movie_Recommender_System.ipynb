[{"id": 28, "name": "Action"}, {"id": 12, "name": "Adventure"}, {"id": 14, "name": "Fantasy"}, {"id": 878, "name": "Science Fiction"}]

 we will convert the above format to the following format:

['Action','Adventure','Fantasy','Science Friction']

```python
def convert(obj):
  L = []
  for i in obj:
    L.append(i['name'])
  return L
```

```python
#convert('[{"id": 28, "name": "Action"}, {"id": 12, "name": "Adventure"}, {"id": 14, "name": "Fantasy"}, {"id": 878, "name": "Science Fiction"}]')
```

This will not work. It asks to convert the string to list. For to do this we have a module in python called 'ast'

```python
import ast
ast.literal_eval('[{"id": 28, "name": "Action"}, {"id": 12, "name": "Adventure"}, {"id": 14, "name": "Fantasy"}, {"id": 878, "name": "Science Fiction"}]')
```

Now we write the convert function using the above module

```python
def convert(obj):
  L = []
  for i in ast.literal_eval(obj):
    L.append(i['name'])
  return L
```

```python
movies['genres'] = movies['genres'].apply(convert)
```

```python
movies.head()
```

We have sorted it for one  column. Now we do the same thing to the keywords column.

```python
movies['keywords'] = movies['keywords'].apply(convert)
```

```python
movies.head()
```

Now we do for cast

For cast we will consider only first 3 actors.

```python
def convert3(obj):
  L = []
  count =0
  for i in ast.literal_eval(obj):
    if count != 3:
      L.append(i['name'])
      count+=1
    else:
      break
  return L
```

```python
movies['cast'] = movies['cast'].apply(convert3)
```

```python
movies.head()
```

Now for crew:
We will consider only the director

```python
movies['crew'][0]
```

```python
def fetch_director(obj):
  L = []
  for i in ast.literal_eval(obj):
    if i['job'] == 'Director':
      L.append(i['name'])
  return L
```

```python
movies['crew'] = movies['crew'].apply(fetch_director)
```

```python
movies.head()
```

```python
movies['overview'][0]
```

Overview here is a string. We conver this also to the sting so that we can concatinate with other columns.

```python
movies['overview'] = movies['overview'].apply(lambda x:x.split())
```

```python
movies.head()
```

Now we got anothe thing to do befor concatnating. If there are spaces in the names of cast or crew, we need to remove that.

Ex: Sam Worthington  --->  SamWorthington


```python
movies['genres'] = movies['genres'].apply(lambda x:[i.replace(" ","") for i in x])
```

```python
movies['keywords'] = movies['keywords'].apply(lambda x:[i.replace(" ","") for i in x])
movies['cast'] = movies['cast'].apply(lambda x:[i.replace(" ","") for i in x])
movies['crew'] = movies['crew'].apply(lambda x:[i.replace(" ","") for i in x])
```

```python
movies.head()
```

Now its time to concatinate the columns

```python
movies['tags'] = movies['overview'] + movies['genres'] + movies['keywords'] + movies['cast'] + movies['crew']
```

```python
movies.head()
```

```python
new_df = movies[['id','title','tags']]
```

```python
new_df.head()
```

Now we convert the tags from list to string by separating them at commas

```python
new_df['tags'] = new_df['tags'].apply(lambda x:" ".join(x))
```

```python
new_df.head()
```

It is suggested to convert all into lower case letters

```python
new_df['tags'] = new_df['tags'].apply(lambda x:x.lower())
```

```python
new_df.head()
```

```python
import nltk
```

```python
from nltk.stem.porter import PorterStemmer
ps = PorterStemmer()
```

```python
def stem(text):
  y = []

  for i in text.split():
    y.append(ps.stem(i))

  return " ".join(y)
```

```python
new_df['tags'] = new_df['tags'].apply(stem)
```

```python
new_df['tags'][0]
```

```python
new_df['tags'][1]
```

Now using this data, we need to find a way to find similar movies.

One such way is by finding no.of similar words and comparing it with remaining movies. Honestely speaking, it is not a good idea.

So, Here comes the need of **Vectorization**.    
we will convert total text to vector.

```python
from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer(max_features = 5000,stop_words='english')
```

```python
vectors = cv.fit_transform(new_df['tags']).toarray()
```

```python
vectors[0]
```

```python
cv.get_feature_names_out()
```

We apply steming here. It removes the similar words

Ex: ['loved','loving','love'] ---> ['love','love','love']

```python
ps.stem('loving')
```

```python
stem('in the 22nd century, a paraplegic marine is dispatched to the moon pandora on a unique mission, but becomes torn between following orders and protecting an alien civilization. action adventure fantasy sciencefiction cultureclash future spacewar spacecolony society spacetravel futuristic romance space alien tribe alienplanet cgi marine soldier battle loveaffair antiwar powerrelations mindandsoul 3d samworthington zoesaldana sigourneyweaver jamescameron')
```

Now we got the vectors of every movie. Now we find the distance of every movie with other movie.

Note that here we donot consider euclidian distance instead we consider cosine distance(i.e., the angle between movies).

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
cosine_similarity(vectors).shape
```

```python
similarity = cosine_similarity(vectors)
```

```python
similarity[0]
```

We need to sort the distanes so that similar movies get together. With sorting we get a problem that is, we loose the index of movies. We need to keep the track of indices. For to do that we use 'enumerate' function.

```python
enumerate(similarity[0])
```

```python
sorted(list(enumerate(similarity[0])),reverse=True, key = lambda x:x[1])[1:6]
```

```python
def recommend(movie):
  movie_index = new_df[new_df['title'] == movie].index[0]
  distances = similarity[movie_index]
  movies_list = sorted(list(enumerate(distances)),reverse=True, key = lambda x:x[1])[1:6]

  for i in movies_list:
    print(new_df.iloc[i[0]].title)
```

```python
recommend('Avatar')
```

```python
recommend('Batman Begins')
```
